```python
import requests
import random

# Your API Key
api_key = "GQ2IN18GPU7JN2YD"  # Replace with your API key

# List of currency pairs or other assets
assets = ["EUR/USD", "USD/JPY", "GBP/USD"]

# Function to format assets for Alpha Vantage (e.g., EUR/USD -> EURUSD)
def format_asset(asset):
    return asset.replace("/", "")

# Function to create randomized trading signals
def generate_signal(opening_price, asset):
    # Random exit percentage between 0.5% to 1.5%
    buy_exit_percentage = random.uniform(0.5, 1.5) / 100
    sell_exit_percentage = random.uniform(0.5, 1.5) / 100

    buy_exit_price = opening_price * (1 + buy_exit_percentage)
    sell_exit_price = opening_price * (1 - sell_exit_percentage)

    return f"{asset} - Buy Exit: {buy_exit_price:.4f} (at +{buy_exit_percentage*100:.2f}%), Sell Exit: {sell_exit_price:.4f} (at -{sell_exit_percentage*100:.2f}%)"

# Loop through each asset to fetch the opening price and generate signals
for asset in assets:
    from_symbol, to_symbol = asset.split("/")
    url = f"https://www.alphavantage.co/query?function=FX_DAILY&from_symbol={from_symbol}&to_symbol={to_symbol}&apikey={'GQ2IN18GPU7JN2YD'}"

    # Fetch data from the API
    response = requests.get(url)
    data = response.json()

    # Check and display results
    if "Time Series FX (Daily)" in data:
        latest_date = list(data["Time Series FX (Daily)"].keys())[0]
        opening_price = float(data["Time Series FX (Daily)"][latest_date]["1. open"])

        # Generate and print the signal
        signal = generate_signal(opening_price, asset)
        print(signal)  # This will output text only for each asset
    else:
        print(f"Error retrieving data for {asset}")
```
